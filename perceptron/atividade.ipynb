```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Importing things, Reshaping stuff and Createing Weights
X, y = make_blobs(centers=2, cluster_std=1, random_state=1)
y = np.where(y == 0, -1, y)

xr = X[:, np.newaxis, :]
newX = np.concatenate((xr, np.ones((X.shape[0], 1, 1))), axis=2)

rng = np.random
w = rng.random(3)
print(f"Inicial W: {w}")

LEARNING = 0.2

def sign(entrada):
    
    return 1 if entrada>0 else -1

def update(w_old,learning,label,y_pred,x):
    return w_old+(LEARNING*(label-y_pred)*x)

def perceptron(newX,w,y):
    for time in range(20):
        for i,x in enumerate(newX):
            label = y[i]
            peso_atual = w
            x=x[0]
            dot = np.dot(peso_atual,x)
            y_pred = sign(dot)
            novopeso = update(peso_atual,LEARNING,label,y_pred,x)
            w = novopeso

def plot(w):
    print(f"Final: {w}")
    plt.scatter(X[:,0], X[:,1], c=y)

    w1, w2, w3 = w

    x_line = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), num=100)
    y_line = (-w1 * x_line - w3) / w2

    plt.scatter(X[:, 0], X[:, 1], c=y)

    plt.plot(x_line, y_line, color='r')

    plt.show()
print(f"final: {w}")
perceptron(newX,w,y)
plot(w)
```
